### 1. Gr√°fico de Linha (Line Plot)

üìñ **Explica√ß√£o Conceitual**
O gr√°fico de linha √© ideal para visualizar a evolu√ß√£o de uma vari√°vel cont√≠nua ao longo do tempo ou de uma sequ√™ncia. Ele conecta pontos de dados com uma linha, facilitando a identifica√ß√£o de tend√™ncias, picos e vales.

* **Par√¢metros Principais (`plt.plot`):**
    * `x`, `y`: As coordenadas dos pontos. `x` geralmente representa o tempo ou a sequ√™ncia, e `y` o valor observado.
    * `marker`: Estilo do marcador para cada ponto (ex: 'o' para c√≠rculo).
    * `linestyle`: Estilo da linha (ex: '--' para tracejada).
    * `color`: Cor da linha.
    * `label`: Texto para a legenda do gr√°fico.

üìä **Dados Fixos**
Vendas mensais de uma pequena livraria em 2024 (em milhares de R$).

```python
# --- Dados Fixos ---
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas_2024 = [35, 37, 42, 40, 45, 55, 52, 58, 60, 55, 62, 70]

# --- Cria√ß√£o do Gr√°fico ---
# Define o tamanho da figura para melhor visualiza√ß√£o
plt.figure(figsize=(10, 5))

# Plota os dados de vendas ao longo dos meses
plt.plot(meses, vendas_2024, marker='o', linestyle='-', color='b', label='Vendas (em milhares de R$)')

# --- Melhorando a Apresenta√ß√£o ---
# Adiciona um t√≠tulo ao gr√°fico
plt.title('Evolu√ß√£o das Vendas Mensais da Livraria em 2024', fontsize=16)

# Adiciona r√≥tulos aos eixos X e Y
plt.xlabel('M√™s', fontsize=12)
plt.ylabel('Vendas (em milhares de R$)', fontsize=12)

# Adiciona uma anota√ß√£o para destacar o melhor m√™s
pico_vendas = max(vendas_2024)
mes_pico = meses[vendas_2024.index(pico_vendas)]
plt.annotate(f'Pico de Vendas\n{mes_pico}: R${pico_vendas}k',
             xy=(mes_pico, pico_vendas),
             xytext=(mes_pico, pico_vendas + 5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

# Habilita a legenda
plt.legend()

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O gr√°fico mostra uma clara tend√™ncia de crescimento nas vendas ao longo de 2024. O ano come√ßa com vendas de `R$`35k em janeiro e termina com um pico de `R$`70k em dezembro, provavelmente impulsionado pelas festas de fim de ano. H√° uma pequena queda em abril e julho, mas a recupera√ß√£o √© r√°pida. O segundo semestre foi significativamente mais forte que o primeiro, indicando uma acelera√ß√£o positiva nos neg√≥cios.

### 2. Gr√°fico de Barras Verticais (Vertical Bar Chart)

üìñ **Explica√ß√£o Conceitual**
O gr√°fico de barras verticais √© usado para comparar quantidades entre diferentes categorias. A altura de cada barra √© proporcional ao valor que ela representa, tornando as compara√ß√µes diretas e intuitivas.

* **Par√¢metros Principais (`plt.bar`):**
    * `x`: As categorias no eixo horizontal.
    * `height`: Os valores num√©ricos que determinam a altura das barras.
    * `width`: A largura das barras (padr√£o 0.8).
    * `color`: Cor das barras.

üìä **Dados Fixos**
N√∫mero de inscritos em diferentes cursos de uma plataforma online.

```python
# --- Dados Fixos ---
cursos = ['Python para Dados', 'SQL Essencial', 'Machine Learning', 'Cloud Computing', 'Marketing Digital']
inscritos = [1200, 950, 780, 620, 1150]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 6))

# Cria o gr√°fico de barras
barras = plt.bar(cursos, inscritos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

# --- Melhorando a Apresenta√ß√£o ---
plt.title('N√∫mero de Inscritos por Curso', fontsize=16)
plt.xlabel('Cursos', fontsize=12)
plt.ylabel('N√∫mero de Inscritos', fontsize=12)

# Rotaciona os r√≥tulos do eixo X para evitar sobreposi√ß√£o
plt.xticks(rotation=15, ha='right')

# Adiciona o valor exato no topo de cada barra
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2.0, altura, f'{altura}', ha='center', va='bottom')

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O curso "Python para Dados" √© o mais popular, com 1.200 inscritos, seguido de perto por "Marketing Digital" com 1.150. Isso sugere uma alta demanda por habilidades em tecnologia e marketing. "SQL Essencial" tamb√©m tem uma base s√≥lida de alunos (950). Cursos mais especializados como "Machine Learning" (780) e "Cloud Computing" (620) possuem menos inscritos, o que pode refletir um p√∫blico mais nichado ou um n√≠vel de dificuldade maior.

### 3. Gr√°fico de Barras Horizontais (Horizontal Bar Chart)

üìñ **Explica√ß√£o Conceitual**
Funciona como o gr√°fico de barras verticais, mas com os eixos trocados. √â particularmente √∫til quando os nomes das categorias s√£o longos, pois evita a sobreposi√ß√£o de r√≥tulos no eixo.

* **Par√¢metros Principais (`plt.barh`):**
    * `y`: As categorias no eixo vertical.
    * `width`: Os valores num√©ricos que determinam o comprimento das barras.
    * `height`: A altura das barras (padr√£o 0.8).

üìä **Dados Fixos**
Tempo m√©dio (em dias) para a conclus√£o de diferentes fases de um projeto de software.

```python
# --- Dados Fixos ---
fases_projeto = ['Planejamento e Requisitos', 'Design de UI/UX', 'Desenvolvimento (Back-end)', 'Desenvolvimento (Front-end)', 'Testes e QA', 'Implanta√ß√£o (Deploy)']
dias_conclusao = [15, 12, 35, 30, 20, 5]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 7))

# Cria o gr√°fico de barras horizontais
barras_h = plt.barh(fases_projeto, dias_conclusao, color='teal')

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Tempo M√©dio de Conclus√£o por Fase do Projeto', fontsize=16)
plt.xlabel('Dias', fontsize=12)
plt.ylabel('Fase do Projeto', fontsize=12)

# Inverte o eixo Y para que a primeira fase apare√ßa no topo
plt.gca().invert_yaxis()

# Adiciona o valor exato na ponta de cada barra
for index, value in enumerate(dias_conclusao):
    plt.text(value, index, f' {value} dias', va='center')

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
As fases de desenvolvimento (Back-end e Front-end) s√£o as que consomem mais tempo, com 35 e 30 dias respectivamente, o que √© esperado em projetos de software. A fase de testes (20 dias) tamb√©m √© um componente temporal significativo, destacando sua import√¢ncia. Planejamento (15 dias) e Design (12 dias) s√£o etapas iniciais mais curtas. A implanta√ß√£o (Deploy) √© a fase mais r√°pida, com apenas 5 dias. O gr√°fico claramente aponta que o "cora√ß√£o" do projeto est√° na codifica√ß√£o e nos testes.

### 4. Histograma (Histogram)

üìñ **Explica√ß√£o Conceitual**
O histograma √© usado para visualizar a distribui√ß√£o de uma √∫nica vari√°vel num√©rica. Ele agrupa os n√∫meros em intervalos (chamados de "bins" ou "caixas") e mostra a frequ√™ncia de ocorr√™ncias em cada intervalo. √â excelente para entender a forma, a tend√™ncia central e a dispers√£o dos dados.

* **Par√¢metros Principais (`plt.hist`):**
    * `x`: A lista de dados num√©ricos.
    * `bins`: O n√∫mero de intervalos para agrupar os dados. Pode ser um n√∫mero inteiro ou uma sequ√™ncia de limites dos intervalos.
    * `edgecolor`: Cor da borda das barras.

üìä **Dados Fixos**
Notas de 50 alunos em uma prova final (de 0 a 100).

```python
# --- Dados Fixos ---
notas_prova = [
    82, 95, 71, 65, 50, 88, 79, 73, 91, 84,
    68, 75, 89, 93, 70, 66, 81, 85, 94, 77,
    55, 62, 76, 83, 90, 98, 72, 69, 80, 87,
    74, 92, 96, 78, 67, 86, 99, 60, 58, 81,
    75, 88, 91, 79, 70, 65, 85, 90, 80, 73
]

# --- C√°lculos Estat√≠sticos ---
media = statistics.mean(notas_prova)
mediana = statistics.median(notas_prova)
desvio_padrao = statistics.stdev(notas_prova)

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 6))

# Cria o histograma com 10 bins
plt.hist(notas_prova, bins=10, edgecolor='black', color='skyblue')

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Distribui√ß√£o das Notas da Prova Final', fontsize=16)
plt.xlabel('Notas', fontsize=12)
plt.ylabel('Frequ√™ncia (N¬∫ de Alunos)', fontsize=12)

# Adiciona linhas verticais para m√©dia e mediana
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'M√©dia: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')

# Adiciona um texto com o desvio padr√£o
plt.text(55, 11, f'Desvio Padr√£o: {desvio_padrao:.2f}', bbox=dict(facecolor='white', alpha=0.5))

plt.legend()
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O histograma mostra que a maior concentra√ß√£o de alunos (frequ√™ncia mais alta) obteve notas na faixa de 80 a 90. A distribui√ß√£o √© levemente assim√©trica √† esquerda (negativa), com uma "cauda" de notas mais baixas. A m√©dia (80.14) √© ligeiramente menor que a mediana (81.0), o que confirma essa leve assimetria. Poucos alunos tiraram notas abaixo de 60. A maioria da turma teve um bom desempenho, concentrando-se acima da nota 70.

### 5. Gr√°fico de Dispers√£o (Scatter Plot)

üìñ **Explica√ß√£o Conceitual**
O gr√°fico de dispers√£o (ou scatter plot) exibe a rela√ß√£o entre duas vari√°veis num√©ricas. Cada ponto no gr√°fico representa uma observa√ß√£o, com sua posi√ß√£o determinada pelos valores das duas vari√°veis (uma no eixo X, outra no eixo Y). √â fundamental para identificar correla√ß√µes.

* **Par√¢metros Principais (`plt.scatter`):**
    * `x`, `y`: As duas listas de dados num√©ricos de mesmo tamanho.
    * `s`: Tamanho dos pontos.
    * `c`: Cor dos pontos.
    * `alpha`: Transpar√™ncia dos pontos (√∫til para dados sobrepostos).

üìä **Dados Fixos**
Rela√ß√£o entre horas de estudo por semana e a nota obtida em um exame.

```python
# --- Dados Fixos ---
horas_estudo = [2, 3, 5, 6, 7, 8, 10, 11, 13, 14, 16, 17, 18, 20]
nota_exame =   [55, 60, 62, 68, 70, 75, 74, 80, 82, 88, 85, 91, 95, 98]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 6))

# Cria o gr√°fico de dispers√£o
plt.scatter(horas_estudo, nota_exame, color='purple', s=100, alpha=0.7, edgecolors='black')

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Rela√ß√£o entre Horas de Estudo e Nota no Exame', fontsize=16)
plt.xlabel('Horas de Estudo por Semana', fontsize=12)
plt.ylabel('Nota no Exame (0-100)', fontsize=12)

# Adiciona uma anota√ß√£o para um ponto espec√≠fico
plt.annotate('Poss√≠vel outlier?\nBaixo rendimento',
             xy=(10, 74),
             xytext=(12, 65),
             arrowprops=dict(facecolor='gray', shrink=0.05))

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O gr√°fico mostra uma forte correla√ß√£o positiva entre as horas de estudo e a nota no exame. √Ä medida que as horas de estudo aumentam (eixo X), a nota tende a aumentar tamb√©m (eixo Y). Os pontos formam um padr√£o que se assemelha a uma linha ascendente. H√° um ponto (10 horas, nota 74) que parece estar um pouco abaixo da tend√™ncia geral, sugerindo um rendimento ligeiramente menor para aquela dedica√ß√£o, mas no geral, a rela√ß√£o √© clara e consistente.

### 6. Gr√°fico de Pizza (Pie Chart)

üìñ **Explica√ß√£o Conceitual**
O gr√°fico de pizza (ou de setores) √© usado para representar as propor√ß√µes de diferentes categorias que comp√µem um todo. Cada "fatia" representa uma categoria, e sua √°rea √© proporcional √† porcentagem da categoria no total. √â mais eficaz com poucas categorias.

* **Par√¢metros Principais (`plt.pie`):**
    * `x`: Os valores num√©ricos de cada categoria.
    * `labels`: Os nomes de cada categoria (fatia).
    * `autopct`: Formato para exibir o valor percentual na fatia.
    * `explode`: Uma tupla para "destacar" (separar) uma ou mais fatias.
    * `startangle`: √Çngulo inicial para a primeira fatia.

üìä **Dados Fixos**
Composi√ß√£o do or√ßamento mensal de uma fam√≠lia.

```python
# --- Dados Fixos ---
categorias_orcamento = ['Moradia', 'Alimenta√ß√£o', 'Transporte', 'Educa√ß√£o', 'Lazer', 'Outros']
gastos = [1800, 1200, 500, 800, 400, 300]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(8, 8))

# Define qual fatia ser√° destacada (explode)
explode = (0.05, 0, 0, 0, 0, 0) # Destaca a primeira fatia ('Moradia')

# Cria o gr√°fico de pizza
plt.pie(gastos, labels=categorias_orcamento, autopct='%1.1f%%',
        startangle=90, explode=explode, shadow=True,
        colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'orange', 'lightgray'])

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Distribui√ß√£o do Or√ßamento Mensal', fontsize=16)

# Garante que a pizza seja um c√≠rculo
plt.axis('equal')

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
A "Moradia" √© a maior despesa, consumindo 36% do or√ßamento total, o que √© comum em or√ßamentos familiares. A "Alimenta√ß√£o" vem em segundo lugar, com 24%. Juntas, essas duas categorias representam 60% dos gastos. "Educa√ß√£o" (16%) e "Transporte" (10%) tamb√©m s√£o despesas relevantes. "Lazer" e "Outros" correspondem √†s menores fatias, com 8% e 6% respectivamente. O gr√°fico oferece uma vis√£o imediata de para onde o dinheiro est√° indo.

### 7. Gr√°fico de √Årea Preenchida (Area Chart)

üìñ **Explica√ß√£o Conceitual**
Similar a um gr√°fico de linha, mas a √°rea entre a linha e o eixo √© preenchida com cor, enfatizando o volume ou a magnitude ao longo do tempo. Quando usado com m√∫ltiplas s√©ries de dados, geralmente √© empilhado (`stackplot`), mostrando como a composi√ß√£o de um total muda ao longo do tempo.

* **Par√¢metros Principais (`plt.stackplot`):**
    * `x`: O eixo do tempo ou sequ√™ncia.
    * `y1`, `y2`, ...: As s√©ries de dados a serem empilhadas.
    * `labels`: Os nomes de cada s√©rie de dados.

üìä **Dados Fixos**
Fontes de tr√°fego de um site (Org√¢nico, Social, Direto) ao longo de uma semana.

```python
# --- Dados Fixos ---
dias_semana = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'S√°b', 'Dom']
trafego_organico = [150, 160, 180, 210, 250, 300, 280]
trafego_social =   [80,  85,  90,  100, 110, 150, 160]
trafego_direto =   [50,  55,  60,  65,  70,  80,  75]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 6))

# Cria o gr√°fico de √°rea empilhada
plt.stackplot(dias_semana, trafego_organico, trafego_social, trafego_direto,
              labels=['Org√¢nico', 'Social', 'Direto'],
              colors=['#3498db', '#e74c3c', '#95a5a6'])

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Fontes de Tr√°fego do Site Durante a Semana', fontsize=16)
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('N√∫mero de Visitantes', fontsize=12)
plt.legend(loc='upper left')

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O gr√°fico mostra o volume total de tr√°fego e a contribui√ß√£o de cada fonte ao longo da semana. O tr√°fego total cresce consistentemente, atingindo um pico no fim de semana (S√°bado e Domingo). O tr√°fego "Org√¢nico" (azul) √© a maior fonte durante toda a semana e impulsiona o crescimento geral. O tr√°fego "Social" (vermelho) e "Direto" (cinza) tamb√©m crescem, mas de forma mais modesta. O gr√°fico deixa claro que a estrat√©gia de SEO (tr√°fego org√¢nico) √© a mais eficaz para este site.

### 8. M√∫ltiplas Linhas no Mesmo Gr√°fico

üìñ **Explica√ß√£o Conceitual**
Plotar m√∫ltiplas linhas no mesmo eixo √© uma maneira eficaz de comparar a tend√™ncia de duas ou more s√©ries de dados sobre o mesmo intervalo. Permite a compara√ß√£o direta de padr√µes, cruzamentos e diverg√™ncias.

* **T√©cnica:** Simplesmente chame `plt.plot()` v√°rias vezes antes de `plt.show()`.

üìä **Dados Fixos**
Compara√ß√£o da receita de dois produtos (Produto A e Produto B) ao longo de 6 meses.

```python
# --- Dados Fixos ---
meses_semestre = ['Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
receita_produto_A = [5000, 5200, 5100, 5500, 6200, 7000]
receita_produto_B = [4000, 4800, 5300, 5400, 5100, 4800]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 6))

# Plota a linha para o Produto A
plt.plot(meses_semestre, receita_produto_A, marker='s', label='Produto A') # 's' = square marker

# Plota a linha para o Produto B
plt.plot(meses_semestre, receita_produto_B, marker='^', label='Produto B') # '^' = triangle marker

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Comparativo de Receita: Produto A vs. Produto B', fontsize=16)
plt.xlabel('M√™s', fontsize=12)
plt.ylabel('Receita (R$)', fontsize=12)

# Adiciona anota√ß√£o no ponto de cruzamento
plt.annotate('Cruzamento', xy=('Set', 5100), xytext=('Out', 4500),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             ha='center')

plt.legend()
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O Produto A come√ßa com uma receita maior, mas apresenta um crescimento mais lento at√© setembro. O Produto B, por outro lado, tem um forte crescimento inicial e ultrapassa o Produto A em receita em setembro. No entanto, a partir de outubro, as tend√™ncias se invertem drasticamente: o Produto A dispara em vendas, atingindo R$7.000 em dezembro, enquanto o Produto B entra em decl√≠nio, terminando o semestre com R$4.800. Isso pode indicar uma sazonalidade ou o impacto de uma campanha de marketing focada no Produto A no final do ano.

### 9. Boxplot

üìñ **Explica√ß√£o Conceitual**
O Boxplot (ou diagrama de caixa) √© um m√©todo para representar graficamente a distribui√ß√£o de dados num√©ricos atrav√©s de seus quartis. Ele exibe o m√≠nimo, o primeiro quartil (Q1), a mediana (Q2), o terceiro quartil (Q3) e o m√°ximo. √â extremamente √∫til para comparar distribui√ß√µes entre v√°rias categorias e identificar outliers.

* **Par√¢metros Principais (`plt.boxplot`):**
    * `x`: Uma lista de listas, onde cada lista interna √© um conjunto de dados a ser comparado.
    * `labels`: R√≥tulos para cada "caixa".
    * `patch_artist=True`: Preenche as caixas com cor.

üìä **Dados Fixos**
Alturas (em cm) de jogadores de tr√™s equipes diferentes de basquete.

```python
# --- Dados Fixos ---
equipe_alfa = [198, 201, 203, 195, 199, 205, 197, 200, 210, 185] # Uma equipe com um jogador muito baixo (outlier)

equipes_labels = ['Equipe Alfa']

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 7))

# Calcula a mediana
mediana_alfa = statistics.median(equipe_alfa)

# Cria o boxplot
box = plt.boxplot(equipe_alfa, tick_labels=equipes_labels, patch_artist=True)

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Distribui√ß√£o de Alturas por Equipe de Basquete', fontsize=16)
plt.ylabel('Altura (cm)', fontsize=12)

# Adiciona o valor da mediana como texto acima do boxplot
plt.text(1, mediana_alfa, f'{mediana_alfa:.2f} cm', ha='center', va='bottom')


plt.show()
```

üß† **Interpreta√ß√£o Detalhada**

### 10. Gr√°fico de Barras Empilhadas (Stacked Bar Chart)

üìñ **Explica√ß√£o Conceitual**
O gr√°fico de barras empilhadas √© uma varia√ß√£o do gr√°fico de barras usado para mostrar como um total √© dividido em diferentes partes. Cada barra representa uma categoria principal, e os segmentos coloridos dentro da barra representam subcategorias. √â √≥timo para comparar a composi√ß√£o entre as categorias principais.

* **T√©cnica:** Plotamos a primeira s√©rie de dados com `plt.bar()`. Para as s√©ries seguintes, usamos o par√¢metro `bottom` para indicar que elas devem come√ßar no topo da s√©rie anterior.

üìä **Dados Fixos**
Vendas trimestrais de uma loja de eletr√¥nicos, divididas por categoria de produto (Smartphones, Notebooks, Acess√≥rios).

```python
# --- Dados Fixos ---
trimestres = ['T1 2024', 'T2 2024', 'T3 2024', 'T4 2024']
vendas_smartphones = [150, 170, 210, 280]
vendas_notebooks =   [100, 110, 130, 190]
vendas_acessorios =  [80,  90,  100, 120]

# --- Cria√ß√£o do Gr√°fico ---
plt.figure(figsize=(10, 7))

# Barra base (Smartphones)
plt.bar(trimestres, vendas_smartphones, label='Smartphones', color='#2980b9')

# Barra empilhada (Notebooks), come√ßando no topo dos Smartphones
plt.bar(trimestres, vendas_notebooks, bottom=vendas_smartphones, label='Notebooks', color='#c0392b')

# Barra empilhada (Acess√≥rios), come√ßando no topo da soma de Smartphones e Notebooks
# Calculamos a base somando as duas listas anteriores
base_acessorios = [s + n for s, n in zip(vendas_smartphones, vendas_notebooks)]
plt.bar(trimestres, vendas_acessorios, bottom=base_acessorios, label='Acess√≥rios', color='#f1c40f')

# --- Melhorando a Apresenta√ß√£o ---
plt.title('Vendas Trimestrais por Categoria de Produto', fontsize=16)
plt.xlabel('Trimestre', fontsize=12)
plt.ylabel('Vendas (em milhares de unidades)', fontsize=12)
plt.legend()

# Exibe o gr√°fico
plt.show()
```

üß† **Interpreta√ß√£o Detalhada**
O gr√°fico mostra que as vendas totais aumentam a cada trimestre, com um salto expressivo no T4. "Smartphones" (azul) s√£o consistentemente a categoria de maior venda em todos os trimestres, e seu volume de vendas cresce significativamente ao longo do ano. "Notebooks" (vermelho) e "Acess√≥rios" (amarelo) tamb√©m mostram crescimento, mas sua contribui√ß√£o proporcional para o total parece diminuir ligeiramente no T4, pois o crescimento dos smartphones √© mais acentuado. O T4 √© claramente o per√≠odo de vendas mais forte para todas as categorias, provavelmente devido a eventos como Black Friday e Natal.
