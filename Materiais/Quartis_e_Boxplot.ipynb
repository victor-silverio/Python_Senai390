```python
import statistics as stat
import matplotlib.pyplot as plt
```

```python
# Nosso conjunto de dados de notas dos alunos
notas = [78, 85, 62, 95, 88, 72, 81, 90, 75, 83, 68, 92, 86, 79, 80]

# Calculando os quartis
quartis = stat.quantiles(notas, n=4)

q1 = quartis[0]
q2 = quartis[1] # Mediana
q3 = quartis[2]

# Calculando a Amplitude Interquartil (AIQ)
aiq = q3 - q1

print(f"Primeiro Quartil (Q1): {q1}")
print(f"Segundo Quartil (Mediana - Q2): {q2}")
print(f"Terceiro Quartil (Q3): {q3}")
print(f"Amplitude Interquartil (AIQ): {aiq}")

# Criando o Boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(notas, vert=False)
plt.title('Boxplot das Notas dos Alunos')
plt.xlabel('Notas')
plt.grid(True)
plt.show()
```

```python
notas = [78, 85, 45, 100, 88, 60, 81, 98, 75, 83, 55, 92, 86, 79, 80, 22]

# Calculando os quartis
quartis = stat.quantiles(notas, n=4)

q1 = quartis[0]
q2 = quartis[1] # Mediana
q3 = quartis[2]

# Calculando a Amplitude Interquartil (AIQ)
aiq = q3 - q1

print(f"Primeiro Quartil (Q1): {q1}")
print(f"Segundo Quartil (Mediana - Q2): {q2}")
print(f"Terceiro Quartil (Q3): {q3}")
print(f"Amplitude Interquartil (AIQ): {aiq}")

# Criando o Boxplot
plt.figure(figsize=(10, 6))
# patch_artist=True permite preencher a caixa com cor
plt.boxplot(notas, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot das Notas dos Alunos (Dados Dispersos com Outlier)')
plt.xlabel('Notas')
plt.grid(True)
plt.show()
```
