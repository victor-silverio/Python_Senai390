```python
import pandas as pd
import ast
```

```python
df = pd.DataFrame({
    "filme": ["Filme A", "Filme B"],
    "generos": ['["ação", "romance"]', '["comédia", "drama"]']
})
print(df)
```

```python
# Isso que faltou fazermos na última aula, essa linha converte para list
df["generos"] = df["generos"].apply(ast.literal_eval)
```

```python
type(df['generos'][0])
```

```python
df_explodido = df.explode("generos")
df_explodido.reset_index(inplace=True)
df_explodido.drop(columns='index', inplace=True)
df_explodido
```
